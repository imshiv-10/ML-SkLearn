```python
#  y = wx +b
#  charges = age x +b
#  estimate the charges function
def estimator_charges(age, _coefficient, bias):
  return age*_coefficient + bias
```

```python
age , _coefficient, bias = 24, 200 , -40 
print('The user age has {}, he has to pay for insurance per year {}$'.format(age,estimator_charges(age , _coefficient, bias)))
```

```python
estimated_charges_df = estimator_charges(non_smoker_df.age, _coefficient, bias)
```

```python
estimated_charges_df
```

```python
plt.plot(medical_charges_df.age , estimated_charges_df, 'r-o');
```

```python
plt.figure(figsize=(15,8))
target = non_smoker_df.charges
plt.plot(non_smoker_df.age, estimated_charges_df, 'r-', alpha=0.9)
plt.scatter(non_smoker_df.age, non_smoker_df.charges, alpha =0.8)
#non_smoker_df.age.shape,  estimated_charges_df.shape
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Age vs chahrgessss')
plt.legend(['Estimator','Actual']);
```

```python
def try_parameters(w,b):
  plt.figure(figsize=(15,8))
  target = non_smoker_df.charges
  estimated_charges_df = estimator_charges(non_smoker_df.age, w,b)
  plt.plot(non_smoker_df.age, estimated_charges_df, 'r-', alpha=0.9)
  plt.scatter(non_smoker_df.age, target, alpha =0.8)
  #non_smoker_df.age.shape,  estimated_charges_df.shape
  plt.xlabel('Age')
  plt.ylabel('Charges')
  plt.title('Age vs chahrgessss')
  plt.legend(['Estimator','Actual']);
```
