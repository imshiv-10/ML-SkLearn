```python
#  y = wx +b
#  charges = age x +b
#  estimate the charges function
def estimator_charges(age, _coefficient, bias):
  return age*_coefficient + bias
```

```python
age , _coefficient, bias = 24, 200 , -40 
print('The user age has {}, he has to pay for insurance per year {}$'.format(age,estimator_charges(age , _coefficient, bias)))
```
